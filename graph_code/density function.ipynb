```python
import pandas as pd
import numpy
import matplotlib.pyplot as plt


data = pd.read_csv("sDeaths2020.csv")

data.head()
#sns.set(style="whitegrid")
```

```python
data.Age.plot.density(color='green')
plt.title('Density plot for Age(Suicides 2020)')
plt.show()
plt.savefig('densityAge2020.png')
```

```python
data.Race.plot.density(color='blue')
plt.title('Density plot for Race(Sucides 2020)')
plt.show()
plt.savefig('densityRace2020.png')
```

```python
data.Gender.plot.density(color='red')
plt.title('Density plot for Gender(Suicides 2020)')
plt.show()
plt.savefig('densityGender2020.png')
```
