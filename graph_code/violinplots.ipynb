```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import seaborn as sns
data1 = pd.read_csv("sDeaths2019.csv")

data1.head()
```

```python
age = data1.Age
race = data1.Race
gender = data1.Gender

sns.violinplot(x=race,y=gender,data =data1)
plt.show()
plt.savefig('violinplot(2019)(RVG).png')
```

```python
data2 = pd.read_csv("sDeaths2020.csv")
age = data2.Age
race = data2.Race
gender = data2.Gender
sns.violinplot(x=race,y=gender,data =data2, )
plt.show()
plt.savefig('violinplot(2020)(RVG).png')




```

```python
data2 = pd.read_csv("sDeaths2020.csv")
age = data2.Age
race = data2.Race
gender = data2.Gender
sns.violinplot(x=gender,y=age,data =data2,split=True,
                    scale="count", inner="quartile" )
plt.show()
plt.savefig('violinplot(2020)(GVA).png')

```
