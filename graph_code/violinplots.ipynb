```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import seaborn as sns
data1 = pd.read_csv("sDeaths2019.csv")

data1.head()
```

```python
age = data1.Age
race = data1.Race
gender = data1.Gender

sns.violinplot(x=race,y=gender,data =data1)
plt.savefig('violinplot(2019)(RVG).png')
plt.show()
```
