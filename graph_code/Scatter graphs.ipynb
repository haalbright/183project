```python
import pandas as pd
import numpy
import matplotlib.pyplot as plt

data = pd.read_csv("sDeaths2020.csv")
data.head()
```

```python
x = data['Age']
y = data['Total_cases']
plt.scatter(x,y,color='y')
plt.plot(x,y)
plt.title("Age vs Total_cases")
plt.xlabel("Age")
plt.ylabel("Total_cases")
```

```python
x = data['Total_cases']
y = data['Gender']
plt.scatter(x,y,color='g')
plt.plot(x,y)
plt.title("Gender vs Total_cases")
plt.xlabel("Total_cases")
plt.ylabel("Gender")
```

```python
x = data['Total_cases']
y = data['Race']
plt.scatter(x,y,color='b')
plt.plot(x,y)
plt.title("Race vs Total_cases")
plt.xlabel("Total_cases")
plt.ylabel("Race")
```
