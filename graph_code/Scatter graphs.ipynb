```python
import pandas as pd
import numpy
import matplotlib.pyplot as plt

data = pd.read_csv("sDeaths2020.csv")
```

```python
x = data['Age']
y = data['Total_cases']
plt.scatter(x,y,color='y')
plt.title("Age vs Total_cases")
plt.xlabel("Age")
plt.ylabel("Total_cases")
plt.show()
```

```python
x = data['Gender']
y = data['Total_cases']
plt.scatter(x,y,color='g')
plt.title("Gender vs Total_cases")
plt.xlabel("Gender")
plt.ylabel("Total_cases")
plt.show()
```

```python
x = data['Race']
y = data['Total_cases']
plt.scatter(x,y,color='b')
plt.title("Race vs Total_cases")
plt.xlabel("Race")
plt.ylabel("Total_cases")
plt.show()
```
