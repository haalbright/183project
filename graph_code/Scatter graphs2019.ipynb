```python
import pandas as pd
import numpy
import matplotlib.pyplot as plt

data = pd.read_csv("sDeaths2019.csv")
```

```python
x = data['Age']
y = data['Gender']
plt.scatter(x,y,color='y')
plt.title("Age vs Gender")
plt.xlabel("Age")
plt.ylabel("Gender")
plt.show()
```

```python
x = data['Age']
y = data['Race']
plt.scatter(x,y,color='g')
plt.title("Age vs Race")
plt.xlabel("Age")
plt.ylabel("Race")
plt.show()
```
