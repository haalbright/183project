```python

import pandas as pd
import numpy
import matplotlib.pyplot as plt


data = pd.read_csv("sDeaths2019.csv")

data.head()
data['Gender'].plot(kind='hist');
plt.xlabel("Suicide Deaths Based on Gender (2019)")
plt.savefig('histogramG(2019).png')
plt.show()
```
